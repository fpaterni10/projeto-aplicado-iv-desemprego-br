# **Referencial Teórico**

As séries temporais constituem uma classe de dados onde a dependência temporal é o principal fator de análise, permitindo identificar tendências, ciclos e padrões sazonais (ENDERS, 2015). Os modelos tradicionais para lidar com esse tipo de dado incluem o ARIMA (AutoRegressive Integrated Moving Average), que combina componentes autoregressivos (AR), de integração (I) e de médias móveis (MA). Entretanto, quando as séries apresentam padrões sazonais, é necessário recorrer a modelos estendidos como o SARIMA, que incorpora a sazonalidade, e o SARIMAX, que permite a inclusão de variáveis exógenas relevantes, como indicadores macroeconômicos (BEIRÃO; GONÇALVES; NETO, 2021).

Um ponto central em séries temporais é a estacionariedade, que ocorre quando a média e a variância permanecem constantes ao longo do tempo. Séries com forte tendência ou sazonalidade não são estacionárias e precisam ser transformadas, normalmente por diferenciação ou aplicação de logaritmos. Isso é fundamental para que métodos como ARIMA e SARIMA sejam aplicáveis de forma adequada (ENDERS, 2015).

No contexto do desemprego, diversos trabalhos no Brasil têm demonstrado a relevância da modelagem estatística. Becker (2010), em estudo sobre a taxa de desemprego da Região Metropolitana de Porto Alegre, avaliou a aplicação de modelos como Holt-Winters e SARIMA, concluindo que modelos sazonais oferecem melhor ajuste em séries com ciclos econômicos regulares. Já Beirão, Gonçalves e Neto (2021), ao analisar a taxa de desemprego brasileira com a metodologia Box-Jenkins, confirmaram a necessidade de diferenciação para lidar com a não estacionariedade da série.

Além dos métodos estatísticos clássicos, há uma crescente utilização de algoritmos de aprendizado de máquina. Trabalhos recentes apontam que abordagens híbridas, combinando modelos de séries temporais com machine learning, apresentam maior robustez e menor erro de previsão. Um exemplo é o estudo publicado em Estudos Econômicos (2020), que utilizou modelos fatoriais de alta dimensão aliados a técnicas de seleção de variáveis para prever indicadores macroeconômicos, incluindo a taxa de desemprego, alcançando ganhos significativos de precisão.

A literatura mostra, portanto, que a modelagem de séries temporais voltada para a previsão de indicadores de mercado de trabalho no Brasil é um tema relevante, com aplicações práticas em políticas públicas e gestão econômica. O uso combinado de técnicas estatísticas clássicas e métodos modernos de aprendizado de máquina surge como alternativa promissora, unindo a interpretabilidade dos primeiros à flexibilidade dos segundos.

#**EDA e Pré-processamento dos dados**

A análise exploratória teve como objetivo compreender a estrutura temporal das variáveis e identificar padrões de tendência, sazonalidade e persistência autocorrelativa relevantes para a modelagem da taxa de desocupação brasileira.



```python
# =======================
# Carregar CSV e recriar df_model
# =======================
import os, io, requests
import numpy as np, pandas as pd
from pathlib import Path

CSV_LOCAL = Path("dataset/tratados/caged_pnad_selic_trimestral.csv")
RAW_URL = ("https://raw.githubusercontent.com/"
           "fpaterni10/projeto-aplicado-iv-desemprego-br/main/"
           "dataset/tratados/caged_pnad_selic_trimestral.csv")

if CSV_LOCAL.exists():
    df0 = pd.read_csv(CSV_LOCAL)
else:
    r = requests.get(RAW_URL, headers={"User-Agent": "Mozilla/5.0"})
    r.raise_for_status()
    df0 = pd.read_csv(io.StringIO(r.text))

df0["periodo_tri"] = df0["periodo_tri"].astype(str)
df_model = df0.set_index(pd.PeriodIndex(df0["periodo_tri"], freq="Q-DEC")).sort_index()
df_model.index.name = "periodo"
df_model["data"] = df_model.index.to_timestamp(how="end").normalize()

df_model["caged_roll3"] = df_model["caged_saldo_tri"].rolling(3, min_periods=3).mean()
df_model["caged_roll3_asinh"] = np.arcsinh(df_model["caged_roll3"])
df_model["quarter"] = df_model.index.quarter

Q = pd.get_dummies(df_model["quarter"], prefix="Q", drop_first=False).astype("uint8")
df_model["post_2021"] = (df_model["data"] >= "2021-01-01").astype("uint8")

df_model = pd.concat([df_model, Q], axis=1).dropna().copy()

def serie_q_dec(s):
    """Retorna a série trimestral (Q-DEC) com DatetimeIndex no fim do trimestre,
    funcione s.index sendo PeriodIndex ou DatetimeIndex."""
    if isinstance(s.index, pd.PeriodIndex):
        return s.asfreq("Q-DEC").to_timestamp(how="end")
    elif isinstance(s.index, pd.DatetimeIndex):
        return s.to_period("Q-DEC").to_timestamp(how="end")
    else:
        raise TypeError("Índice temporal não é PeriodIndex nem DatetimeIndex.")


print("Dimensão df_model:", df_model.shape)
display(df_model.head())
```

**Contexto da EDA.** Base trimestral 2012Q3–2025Q2 com 52 observações; Q‑DEC como frequência explícita e data no fim do trimestre. Features derivadas: caged_roll3, caged_roll3_asinh, dummies Q_1..Q_4 e post_2021. Esta seção investiga tendência, sazonalidade e persistência para fundamentar a modelagem.

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm

def _acf(series, name, lags=12):
    fig = plt.figure()
    ax = plt.gca()
    sm.graphics.tsa.plot_acf(series.astype(float).values, lags=lags, ax=ax)
    ax.set_title(f"ACF - {name}")
    fig.tight_layout()
    plt.show()
    plt.close(fig)

def _pacf(series, name, lags=12, method="ywm"):
    fig = plt.figure()
    ax = plt.gca()
    sm.graphics.tsa.plot_pacf(series.astype(float).values, lags=lags, method=method, ax=ax)
    ax.set_title(f"PACF - {name}")
    fig.tight_layout()
    plt.show()
    plt.close(fig)
```

## **Autocorrelação (ACF) e Autocorrelação Parcial (PACF)**

### **Taxa de desocupação (PNAD)**

Os gráficos de ACF e PACF mostram decadência lenta das autocorrelações positivas até aproximadamente o 8º lag, indicando alta persistência e memória longa da série, característica de não estacionariedade.
A PACF apresenta pico significativo no primeiro lag e valores residuais pequenos nos demais, sugerindo que um modelo AR de baixa ordem (p≈1) pode capturar parte da dependência temporal.
Esse comportamento reforça a necessidade de diferenciação (d = 1) para tornar a série estacionária antes da modelagem ARIMA/SARIMA.

### **Saldo do CAGED**

A ACF do saldo trimestral mostra autocorrelação moderada em lags curtos e queda rápida após o lag 4, sugerindo estacionariedade parcial.
A PACF indica pico no primeiro lag e pequenas oscilações posteriores, o que aponta para uma estrutura AR(1) ou ARMA(1, 1).

Ao aplicar a média móvel de 3 trimestres (caged_roll3), observa-se suavização da volatilidade e redução das autocorrelações de média duração.
A transformação asinh log-like (caged_roll3_asinh) reduz a amplitude das autocorrelações positivas nos primeiros lags, estabilizando a variância, o que indica melhor comportamento para uso como variável exógena no modelo SARIMAX.

# **Modelo base**


## **Modelo LGBM (Machine Learning)**

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
import warnings, logging
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error
from IPython.display import display

TEST_LAST_N  = 4
SEED         = 42

from pathlib import Path
CSV_FALLBACK = Path("dataset/tratados/caged_pnad_selic_trimestral.csv")
if 'df_model' not in globals():
    assert CSV_FALLBACK.exists(), (
        "df_model não está em memória e o CSV não foi encontrado em "
        f"{CSV_FALLBACK}. Suba o arquivo ou crie df_model antes."
    )
    df_model = pd.read_csv(CSV_FALLBACK)
    if "data" in df_model.columns:
        df_model["data"] = pd.to_datetime(df_model["data"])
        df_model = df_model.set_index("data").sort_index()


features_cal = [c for c in ['caged_roll3_asinh','caged_saldo_tri','Q_1','Q_2','Q_3','Q_4','post_2021']
                if c in df_model.columns]
target = "tx_desocupacao"
assert target in df_model.columns, f"Coluna alvo '{target}' ausente no df_model."

dfm = df_model[features_cal + [target]].copy().interpolate(limit_direction="both")

train_lgb = dfm.iloc[:-TEST_LAST_N].copy()
test_lgb  = dfm.iloc[-TEST_LAST_N:].copy()
X_tr = train_lgb[features_cal]
X_te = test_lgb[features_cal]
y_tr = train_lgb[target].to_numpy()
y_te = test_lgb[target].to_numpy()


logging.getLogger("lightgbm").setLevel(logging.ERROR)
warnings.filterwarnings("ignore", message="X does not have valid feature names")

lgbm = LGBMRegressor(
    n_estimators=400, learning_rate=0.05,
    num_leaves=31, max_depth=4,
    min_child_samples=20, subsample=0.8, colsample_bytree=0.8,
    reg_lambda=0.1, random_state=SEED,
    verbosity=-1,
    verbose=-1
)

lgbm.fit(X_tr, y_tr)

pred_tr = lgbm.predict(X_tr)
bias    = (y_tr[-4:] - pred_tr[-4:]).mean()
pred_te = lgbm.predict(X_te) + bias

def _rmse(y_true, y_pred):
    y_true, y_pred = np.asarray(y_true, float), np.asarray(y_pred, float)
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

def _mape(y_true, y_pred):
    y_true = np.asarray(y_true, float).ravel()
    y_pred = np.asarray(y_pred, float).ravel()
    mask = y_true != 0
    return (np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])).mean() * 100

lgbm_metrics = pd.DataFrame(
    {"MAE":[mean_absolute_error(y_te, pred_te)],
     "RMSE":[_rmse(y_te, pred_te)],
     "MAPE (%)":[_mape(y_te, pred_te)]},
    index=["LGBM (calibrado)"]
)
display(lgbm_metrics)

t_holdout = test_lgb.index.to_timestamp() if hasattr(test_lgb.index, "to_timestamp") else test_lgb.index
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(t_holdout, y_te,    label="Real")
ax.plot(t_holdout, pred_te, label="LGBM (calibrado)")
ax.set_title("Real vs Pred — LGBM (calibrado)")
ax.set_xlabel("Tempo"); ax.set_ylabel("Taxa de desocupação (p.p.)")
ax.legend(); fig.tight_layout()
plt.show(); plt.close(fig)

pred_lgb = pred_te
lgbm_metrics = lgbm_metrics
```

**LGBM (calibrado) — resultados no holdout (4 trimestres)**

* **Desempenho:** MAE **1,2747**, RMSE **1,4567** e MAPE **20,51%** (tabela acima).
* **Comportamento da curva:** mesmo com a calibração de viés, as previsões ficaram planas em patamares elevados e não acompanharam a forma da série real no holdout, com superestimação nos dois últimos trimestres.
* **Leitura:** o conjunto enxuto de features reduz sobreajuste, mas a amostra curta e o regime recente parecem limitar a capacidade do LGBM de capturar viradas de curto prazo.
  *Implica*: manter parsimonioso e usar LGBM como baseline de ML, mas a escolha do modelo final deve considerar a comparação com o SARIMAX.

## **Modelo SARIMAX (Estatístico)**

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from scipy.stats import jarque_bera
from IPython.display import display

TEST_LAST_N  = 4
ORDER        = (1,1,2)
SEAS_ORDER   = (1,0,1,4)

def temporal_holdout(df, n_test=4):
    return df.iloc[:-n_test].copy(), df.iloc[-n_test:].copy()

train_sar, test_sar = temporal_holdout(df_model, n_test=TEST_LAST_N)

y_tr = train_sar["tx_desocupacao"].asfreq("Q-DEC")
y_te = test_sar["tx_desocupacao"].asfreq("Q-DEC")
X_tr = train_sar[["caged_saldo_tri"]].asfreq("Q-DEC")
X_te = test_sar[["caged_saldo_tri"]].asfreq("Q-DEC")

sarimax = SARIMAX(
    endog=y_tr, exog=X_tr, order=ORDER, seasonal_order=SEAS_ORDER,
    enforce_stationarity=False, enforce_invertibility=False
).fit(disp=False, maxiter=500)

fcst     = sarimax.get_forecast(steps=len(y_te), exog=X_te)
pred_sar = fcst.predicted_mean

from sklearn.metrics import mean_absolute_error
def _rmse(a,b): a=np.asarray(a,float); b=np.asarray(b,float); return np.sqrt(np.mean((a-b)**2))
def _mape(a,b): a=np.asarray(a,float); b=np.asarray(b,float); m=a!=0; return (np.abs((a[m]-b[m])/a[m])).mean()*100

y_true = np.asarray(y_te.values, float).ravel()
y_pred = np.asarray(pred_sar.values, float).ravel()
sarimax_metrics = pd.DataFrame(
    {"MAE":[mean_absolute_error(y_true, y_pred)],
     "RMSE":[_rmse(y_true, y_pred)],
     "MAPE (%)":[_mape(y_true, y_pred)]},
    index=["SARIMAX baseline"]
)
display(sarimax_metrics)

t_idx = y_te.index
t_holdout = t_idx.to_timestamp() if isinstance(t_idx, pd.PeriodIndex) else t_idx
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(t_holdout, y_te.values,  label="Real")
ax.plot(t_holdout, pred_sar.values, label="SARIMAX")
ax.set_title("Real vs Pred — SARIMAX (baseline)")
ax.set_xlabel("Tempo"); ax.set_ylabel("Taxa de desocupação (p.p.)")
ax.legend(); plt.tight_layout(); plt.show(); plt.close(fig)

resid = sarimax.resid.dropna()
fig2, ax = plt.subplots(1,2, figsize=(9,3.2))
plot_acf(resid.values, lags=16, ax=ax[0]); ax[0].set_title("ACF — Resíduos (SARIMAX)")
plot_pacf(resid.values, lags=16, method="ywm", ax=ax[1]); ax[1].set_title("PACF — Resíduos (SARIMAX)")
plt.tight_layout(); plt.show(); plt.close(fig2)

lb = acorr_ljungbox(resid, lags=[1,4,8,12,16], return_df=True)
display(lb)

jb_stat, jb_p = jarque_bera(resid)
skew = float(pd.Series(resid).skew())
kurt = float(pd.Series(resid).kurtosis())
print(f"Jarque-Bera: estat={jb_stat:.3f}, p={jb_p:.4f}, skew={skew:.3f}, kurt(excesso)={kurt:.3f}")

try:
    comparativo = pd.concat([lgbm_metrics, sarimax_metrics]).round(4)
    display(comparativo)
    fig, ax = plt.subplots(figsize=(7,5))
    ax.plot(t_holdout, y_te.values,      label="Real")
    ax.plot(t_holdout, pred_lgb,         label="LGBM (calibrado)")
    ax.plot(t_holdout, pred_sar.values,  label="SARIMAX")
    ax.set_title("Real vs Pred — Holdout trimestral (últimos 4 trimestres)")
    ax.set_xlabel("Tempo"); ax.set_ylabel("Taxa de desocupação (p.p.)")
    ax.legend(); plt.tight_layout(); plt.show(); plt.close(fig)
except Exception:
    print("Rode a Célula 9 (LGBM) antes desta")
```

**SARIMAX (baseline, exógena = CAGED)**

* **Desempenho:** MAE 0,2891, RMSE 0,3851, MAPE 4,32% — substancialmente melhor que o LGBM no holdout.
* **Aderência visual:** as previsões seguem a forma da série real e reduzem o viés no último trimestre do holdout.
* **Diagnóstico visual:** resíduos sem tendência; ACF dos resíduos dentro da banda (exceto o pico trivial em lag 0).

**Implica**: a especificação sazonal (1,1,2)×(1,0,1)_4 com CAGED como exógena está coerente com a EDA e entrega erro baixo.

**Resíduos — SARIMAX**

Visualmente, ACF/PACF dos resíduos não exibem picos isolados acima do intervalo de confiança para lags > 0. Formalmente, o Ljung‑Box não rejeita a ausência de autocorrelação para lags curtos (1 e 4), mas rejeita para lags médios (8, 12, 16; p < 0,05), indicando dependência remanescente de médio prazo. O Jarque‑Bera rejeita normalidade (p ≈ 0,000), com assimetria ≈ 2,28 e excesso de curtose ≈ 11,28, comportamento típico de séries macroeconômicas com choques. Como ablação, recomenda‑se testar SARIMAX (CAGED+SELIC) e pequeno ajuste de ordens sazonais/ARMA.

**Diagnósticos — ACF e testes formais**

* **ACF dos resíduos:** barras dentro do intervalo de confiança para lags > 0, sem autocorrelações fortes — bom sinal de captação da dinâmica média.
* **Ljung‑Box:** p = 0,160 em lag 4 (não rejeita autocorrelação), mas p = 0,010 (lag 8) e p = 0,007 (lag 12) rejeitam a hipótese nula, indicando dependência remanescente em lags médios.
* **Jarque‑Bera:** rejeita normalidade (valor muito alto), o que é comum em séries macro com choques.

**Implica**: se necessário, testar um refino leve na especificação (ex.: ajustar ordens sazonais/arma) ou incluir SELIC como exógena adicional para checar se reduz a dependência remanescente.

## **Comparação entre Modelos**

* Visualmente, o SARIMAX acompanha a curva real, enquanto o LGBM mantém plateau e superestima no fim do período.
* **Escolha para o baseline:** SARIMAX (com CAGED como exógena) melhor desempenho e comportamento residual mais aceitável.

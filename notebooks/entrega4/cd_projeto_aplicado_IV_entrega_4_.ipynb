# **Referencial Teórico**

As séries temporais constituem uma classe de dados onde a dependência temporal é o principal fator de análise, permitindo identificar tendências, ciclos e padrões sazonais (ENDERS, 2015). Os modelos tradicionais para lidar com esse tipo de dado incluem o ARIMA (AutoRegressive Integrated Moving Average), que combina componentes autoregressivos (AR), de integração (I) e de médias móveis (MA). Entretanto, quando as séries apresentam padrões sazonais, é necessário recorrer a modelos estendidos como o SARIMA, que incorpora a sazonalidade, e o SARIMAX, que permite a inclusão de variáveis exógenas relevantes, como indicadores macroeconômicos (BEIRÃO; GONÇALVES; NETO, 2021).

Um ponto central em séries temporais é a estacionariedade, que ocorre quando a média e a variância permanecem constantes ao longo do tempo. Séries com forte tendência ou sazonalidade não são estacionárias e precisam ser transformadas, normalmente por diferenciação ou aplicação de logaritmos. Isso é fundamental para que métodos como ARIMA e SARIMA sejam aplicáveis de forma adequada (ENDERS, 2015).

No contexto do desemprego, diversos trabalhos no Brasil têm demonstrado a relevância da modelagem estatística. Becker (2010), em estudo sobre a taxa de desemprego da Região Metropolitana de Porto Alegre, avaliou a aplicação de modelos como Holt-Winters e SARIMA, concluindo que modelos sazonais oferecem melhor ajuste em séries com ciclos econômicos regulares. Já Beirão, Gonçalves e Neto (2021), ao analisar a taxa de desemprego brasileira com a metodologia Box-Jenkins, confirmaram a necessidade de diferenciação para lidar com a não estacionariedade da série.

Além dos métodos estatísticos clássicos, há uma crescente utilização de algoritmos de aprendizado de máquina. Trabalhos recentes apontam que abordagens híbridas, combinando modelos de séries temporais com machine learning, apresentam maior robustez e menor erro de previsão. Um exemplo é o estudo publicado em Estudos Econômicos (2020), que utilizou modelos fatoriais de alta dimensão aliados a técnicas de seleção de variáveis para prever indicadores macroeconômicos, incluindo a taxa de desemprego, alcançando ganhos significativos de precisão.

A literatura mostra, portanto, que a modelagem de séries temporais voltada para a previsão de indicadores de mercado de trabalho no Brasil é um tema relevante, com aplicações práticas em políticas públicas e gestão econômica. O uso combinado de técnicas estatísticas clássicas e métodos modernos de aprendizado de máquina surge como alternativa promissora, unindo a interpretabilidade dos primeiros à flexibilidade dos segundos.

#**EDA e Pré-processamento dos dados**

**Objetivo desta etapa:** Entregar a EDA robusta (tendência, sazonalidade, persistência), documentar o pré‑processamento, implementar um modelo base com validação base (holdout e CV rolling‑origin), analisar diagnósticos de resíduos, comparar com baseline(s), e registrar uma nota técnica reprodutível.



```python
# =======================
# Carregar CSV e recriar df_model
# =======================
import os, io, requests
import numpy as np, pandas as pd
from pathlib import Path
import warnings
from statsmodels.tools.sm_exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)

CSV_LOCAL = Path("dataset/tratados/caged_pnad_selic_trimestral.csv")
RAW_URL = ("https://raw.githubusercontent.com/"
           "fpaterni10/projeto-aplicado-iv-desemprego-br/main/"
           "dataset/tratados/caged_pnad_selic_trimestral.csv")

if CSV_LOCAL.exists():
    df0 = pd.read_csv(CSV_LOCAL)
else:
    r = requests.get(RAW_URL, headers={"User-Agent": "Mozilla/5.0"})
    r.raise_for_status()
    df0 = pd.read_csv(io.StringIO(r.text))

df0["periodo_tri"] = df0["periodo_tri"].astype(str)
df_model = df0.set_index(pd.PeriodIndex(df0["periodo_tri"], freq="Q-DEC")).sort_index()
df_model.index.name = "periodo"
df_model["data"] = df_model.index.to_timestamp(how="end").normalize()

df_model["caged_roll3"] = df_model["caged_saldo_tri"].rolling(3, min_periods=3).mean()
df_model["caged_roll3_asinh"] = np.arcsinh(df_model["caged_roll3"])
df_model["quarter"] = df_model.index.quarter

Q = pd.get_dummies(df_model["quarter"], prefix="Q", drop_first=False).astype("uint8")
df_model["post_2021"] = (df_model["data"] >= "2021-01-01").astype("uint8")

df_model = pd.concat([df_model, Q], axis=1).dropna().copy()

def serie_q_dec(s):
    """Retorna a série trimestral (Q-DEC) com DatetimeIndex no fim do trimestre,
    funcione s.index sendo PeriodIndex ou DatetimeIndex."""
    if isinstance(s.index, pd.PeriodIndex):
        return s.asfreq("Q-DEC").to_timestamp(how="end")
    elif isinstance(s.index, pd.DatetimeIndex):
        return s.to_period("Q-DEC").to_timestamp(how="end")
    else:
        raise TypeError("Índice temporal não é PeriodIndex nem DatetimeIndex.")


print("Dimensão df_model:", df_model.shape)
display(df_model.head())
```

## **Pré‑processamento**

- **Checagem de nulos** (alvo e exógenas) e política de imputação (somente exógenas, sem vazamento).  
- **Choques estruturais** de 2020 controlados via `d_covid`.  
- **Padronização de exógenas** sempre com média/desvio do treino.

```python
# Checagem de nulos
display(df_model.isna().sum().to_frame("n_nulos").T)

# Dummy de choque COVID (2020Q2–2021Q1) para controle de outliers estruturais
df_model["d_covid"] = df_model.index.isin(pd.period_range("2020Q2","2021Q1", freq="Q-DEC")).astype(int)
```

**Conclusão da checagem de nulos.**  
A verificação indica ausência de valores ausentes nas variáveis utilizadas, portanto não foi necessária imputação na amostra atual.  
As transformações do CAGED (`roll3`, `asinh`) não deixaram NAs.  

**Política preventiva:** caso surjam nulos em versões futuras, a imputação será aplicada somente nas exógenas, priorizando `ffill/bfill` calculados no conjunto de treino e, como *fallback*, média sazonal por trimestre (estimada no treino), prevenindo vazamento temporal. A variável `d_covid` controla o choque 2020Q2–2021Q1.

## **Decomposição Temporal**

**CAGED (saldo):** volatilidade alta e sazonalidade marcada; queda acentuada em 2020 e retomada parcial 2021–2022. Os resíduos de 2020–2021 são elevados, evidenciando eventos atípicos.


```python
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# PNAD
s = serie_q_dec(df_model["tx_desocupacao"]).dropna()
dec = seasonal_decompose(s, period=4, model="additive", two_sided=False)
fig = dec.plot(); fig.suptitle("Decomposição — tx_desocupacao", y=0.98); fig.tight_layout()
plt.show(); plt.close(fig)
```

Quebra/choque em 2020, com reversão forte em 2021.

**Sazonalidade:** trimestral pronunciada.

**Resíduos:** dispersão elevada em 2020–2021.

**Implicação:** controlar mudança de regime no pós‑2021 (post_2021) e usar CAGED como exógena em SARIMAX.

**Saldo de empregos (CAGED):** revela forte volatilidade e sazonalidade marcada, com picos positivos e negativos recorrentes a cada ano.
A tendência é ligeiramente positiva até 2019, seguida de queda acentuada em 2020, refletindo os impactos da Covid-19, e posterior retomada parcial em 2021-2022.
O componente residual apresenta valores elevados em 2020, indicando a ocorrência de eventos atípicos.

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm

def _acf(series, name, lags=12):
    fig = plt.figure()
    ax = plt.gca()
    sm.graphics.tsa.plot_acf(series.astype(float).values, lags=lags, ax=ax)
    ax.set_title(f"ACF - {name}")
    fig.tight_layout()
    plt.show()
    plt.close(fig)

def _pacf(series, name, lags=12, method="ywm"):
    fig = plt.figure()
    ax = plt.gca()
    sm.graphics.tsa.plot_pacf(series.astype(float).values, lags=lags, method=method, ax=ax)
    ax.set_title(f"PACF - {name}")
    fig.tight_layout()
    plt.show()
    plt.close(fig)
```

## **Autocorrelação (ACF) e Autocorrelação Parcial (PACF)**

### **Taxa de desocupação (PNAD)**

Os gráficos de ACF e PACF da taxa de desocupação evidenciam forte persistência temporal.
A ACF apresenta decaimento lento e gradual, mantendo autocorrelações positivas até aproximadamente o 8º lag, comportamento típico de séries não estacionárias em nível, isto é, o valor atual depende fortemente dos valores passados, e choques tendem a ter efeitos duradouros ao longo do tempo.

A PACF mostra um pico expressivo em lag 1 e valores muito menores nos lags seguintes, o que indica que a dependência direta mais relevante ocorre com o valor imediatamente anterior. Esse padrão é característico de um processo autoregressivo de primeira ordem (AR(1)) sobreposto a um componente integrado (diferenciação necessária).

- A série apresenta memória longa e não estacionariedade.

- A presença de sazonalidade trimestral (picos em múltiplos de 4 na ACF) reforça a necessidade de uma estrutura SARIMA/SARIMAX sazonal.

- Modelagem adequada: usar diferenciação regular (d = 1) e sazonalidade s = 4, com ordem AR pequena (p ≈ 1) para capturar a dependência de curto prazo.

### **Saldo do CAGED**

Na série bruta do CAGED, a ACF mostra autocorrelações altas e oscilatórias em múltiplos de quatro períodos (4, 8, 12…), sinal claro de sazonalidade trimestral persistente. Há também correlações curtas positivas (lags 1–2) e reversões de sinal nos lags médios, típicas de séries com componente cíclico e alto ruído.

A PACF exibe pico relevante no lag 1 e decaimento rápido em seguida, sugerindo uma estrutura AR(1), mas a presença dos picos sazonais indica que o modelo precisa capturar dependência a cada quatro trimestres (P = 1).

Após aplicar a média móvel de 3 trimestres (caged_roll3), observa-se que a ACF passa a decair mais rapidamente, e a PACF perde parte das oscilações, evidenciando redução da variância e do ruído de alta frequência.
A transformação asinh(caged_roll3) suaviza ainda mais as amplitudes, normalizando extremos e mantendo o padrão de sazonalidade, mas com autocorrelações de curta duração.

- O CAGED é parcialmente estacionário, com forte sazonalidade e alta variabilidade.

- Suavização (MM3) e transformação asinh melhoram as propriedades estatísticas (reduzem heterocedasticidade e autocorrelação residual).

- O comportamento da ACF/PACF transformada indica que o CAGED pode ser usado como variável exógena estável em modelos lineares como o SARIMAX, desde que padronizado e preferencialmente sem defasagem simultânea, ou seja, utilizando defasagens X(t–1), X(t–2) conforme verificação da CCF (cross-correlation).

O primeiro obstáculo foi a diferença de granularidade entre as fontes (CAGED mensal e PNAD trimestral), que exigiu o desenvolvimento de uma rotina de agregação temporal coerente (Q-DEC) para alinhar as datas de referência. Esse ajuste evitou desalinhamentos entre trimestres e reduziu ruído em correlações cruzadas.

Outro ponto crítico foi o tratamento dos valores atípicos em 2020–2021, período fortemente impactado pela pandemia. Os gráficos de decomposição e os resíduos indicaram quebra estrutural no comportamento das séries, exigindo a criação de uma variável binária (d_covid) para controlar o efeito de regime e impedir que o modelo interpretasse os picos como tendência permanente.

Além disso, a aplicação inicial da média móvel (roll3) gerou NaN nos dois primeiros trimestres. O problema foi resolvido ajustando o parâmetro min_periods=1 no cálculo da janela, permitindo suavização causal sem perda de observações.

Nas autocorrelações (ACF/PACF), observou-se um decaimento lento na PNAD, caracterizando não-estacionariedade, e picos em múltiplos de quatro no CAGED, revelando sazonalidade trimestral persistente. Essas descobertas orientaram as escolhas da modelagem: diferenciação regular (d = 1) e componente sazonal (s = 4).

Por fim, foi constatada uma correlação contemporânea fraca entre CAGED e PNAD (r ≈ 0,02), o que reforçou a necessidade de testar defasagens do CAGED como exógenas nos modelos subsequentes.
De forma geral, a EDA foi essencial para garantir a integridade do dataset e a coerência das hipóteses de modelagem.

## **Modelo SARIMAX (Estatístico)**

```python
import warnings, numpy as np, pandas as pd, matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.stats.diagnostic import acorr_ljungbox, het_arch
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from scipy.stats import jarque_bera
from sklearn.metrics import mean_absolute_error
from IPython.display import display

TEST_LAST_N  = 4
ORDER        = (1,1,2)
SEAS_ORDER   = (1,0,1,4)   # s=4

def temporal_holdout(df, n_test=4):
    return df.iloc[:-n_test].copy(), df.iloc[-n_test:].copy()

# -------- Split --------
train_sar, test_sar = temporal_holdout(df_model, n_test=TEST_LAST_N)

y_tr = train_sar["tx_desocupacao"].asfreq("Q-DEC")
y_te = test_sar["tx_desocupacao"].asfreq("Q-DEC")
X_tr = train_sar[["caged_saldo_tri"]].asfreq("Q-DEC")
X_te = test_sar[["caged_saldo_tri"]].asfreq("Q-DEC")

# -------- Padronização da exógena --------
mu = X_tr.mean()
sigma = X_tr.std(ddof=0).replace(0, 1.0)
X_tr_std = (X_tr - mu) / sigma
X_te_std = (X_te - mu) / sigma

# -------- Estimação (Powell -> L-BFGS), SEM tol=, com restrições e fallback --------
model = SARIMAX(
    endog=y_tr, exog=X_tr_std,
    order=ORDER, seasonal_order=SEAS_ORDER,
    enforce_stationarity=True, enforce_invertibility=True
)

# 1ª fase: Powell para bom start
res0 = model.fit(method="powell", maxiter=400, disp=False)

# 2ª fase: L-BFGS (sem tol) + cov_type='oim'
sarimax = model.fit(
    start_params=res0.params,
    method="lbfgs",
    maxiter=5000,
    disp=False,
    cov_type="oim"
)

# Fallback: se não convergiu, relaxa restrições para obter start e reimpõe restrições
if hasattr(sarimax, "mle_retvals") and not sarimax.mle_retvals.get("converged", True):
    model_relax = SARIMAX(
        endog=y_tr, exog=X_tr_std,
        order=ORDER, seasonal_order=SEAS_ORDER,
        enforce_stationarity=False, enforce_invertibility=False
    )
    res_relax = model_relax.fit(method="lbfgs", maxiter=3000, disp=False, cov_type="oim")

    model_constr = SARIMAX(
        endog=y_tr, exog=X_tr_std,
        order=ORDER, seasonal_order=SEAS_ORDER,
        enforce_stationarity=True, enforce_invertibility=True
    )
    sarimax = model_constr.fit(method="lbfgs", start_params=res_relax.params,
                               maxiter=5000, disp=False, cov_type="oim")

# -------- Checagem de raízes; se inválidas, tenta simplificações locais (p/q/P/Q-1) --------
def _minabs(z):
    z = np.asarray(z);  return None if z.size == 0 else float(np.min(np.abs(z)))

armin = _minabs(getattr(sarimax, "arroots", np.array([])))
mamin = _minabs(getattr(sarimax, "maroots", np.array([])))

if (armin is not None and armin <= 1.0) or (mamin is not None and mamin <= 1.0):
    p,d,q = ORDER
    P,D,Q,s = SEAS_ORDER
    candidatos = []
    for pp in {max(0,p-1), p}:
        for qq in {max(0,q-1), q}:
            for PP in {max(0,P-1), P}:
                for QQ in {max(0,Q-1), Q}:
                    ord_c = (pp,d,qq); seas_c = (PP,D,QQ,s)
                    try:
                        m = SARIMAX(y_tr, exog=X_tr_std, order=ord_c, seasonal_order=seas_c,
                                    enforce_stationarity=True, enforce_invertibility=True)
                        r = m.fit(method="lbfgs", maxiter=3000, disp=False, cov_type="oim")
                        ar_c = _minabs(getattr(r, "arroots", np.array([])))
                        ma_c = _minabs(getattr(r, "maroots", np.array([])))
                        valid = ((ar_c is None or ar_c > 1.0) and (ma_c is None or ma_c > 1.0))
                        if valid:
                            candidatos.append((r.aic, r, ord_c, seas_c, ar_c, ma_c))
                    except Exception:
                        pass
    if candidatos:
        candidatos.sort(key=lambda x: (x[0], -(x[4] or 9e9), -(x[5] or 9e9)))
        best = candidatos[0]
        sarimax = best[1]
        ORDER, SEAS_ORDER = best[2], best[3]
        armin, mamin = best[4], best[5]

print(">> Especificação final usada:", "ORDER", ORDER, "| SEAS", SEAS_ORDER)
print(f">> min|AR root|={armin}, min|MA root|={mamin} (válido se ambos > 1)")

# -------- Forecast (único, consistente com padronização) --------
fcst = sarimax.get_forecast(steps=len(y_te), exog=X_te_std)
pred_sar = fcst.predicted_mean
ci = fcst.conf_int(alpha=0.05)  # 95%
ci_low = ci.iloc[:, 0].to_numpy()
ci_high = ci.iloc[:, 1].to_numpy()

# -------- Métricas (modelo + baseline sNaive) --------
def _rmse(a,b): a=np.asarray(a,float); b=np.asarray(b,float); return np.sqrt(np.mean((a-b)**2))
def _mape(a,b):
    a=np.asarray(a,float); b=np.asarray(b,float)
    m=a!=0
    return (np.abs((a[m]-b[m])/np.maximum(np.abs(a[m]),1e-12))).mean()*100

y_true = np.asarray(y_te.values, float).ravel()
y_pred = np.asarray(pred_sar.values, float).ravel()

sarimax_metrics = pd.DataFrame(
    {"MAE":[mean_absolute_error(y_true, y_pred)],
     "RMSE":[_rmse(y_true, y_pred)],
     "MAPE (%)":[_mape(y_true, y_pred)]},
    index=["SARIMAX baseline"]
).round(4)
display(sarimax_metrics)

# Baseline sNaive (repete último ciclo sazonal)
def sNaive(y_train, h, s):
    last_season = y_train.iloc[-s:]
    reps = int(np.ceil(h/s))
    return pd.Series(np.tile(last_season.values, reps)[:h], index=y_te.index)

s = SEAS_ORDER[3]
snaive = sNaive(y_tr, len(y_te), s)
rmse_snaive = _rmse(y_true, snaive.values)
print(f">> RMSE sNaive = {rmse_snaive:.3f} | Ganho vs sNaive = {100*(1 - _rmse(y_true,y_pred)/max(rmse_snaive,1e-12)):.1f}%")

# -------- Gráfico Real vs Pred com IC --------
t_idx = y_te.index
t_holdout = t_idx.to_timestamp() if isinstance(t_idx, pd.PeriodIndex) else t_idx
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(t_holdout, y_te.values,          label="Real")
ax.plot(t_holdout, pred_sar.values,      label="SARIMAX")
ax.fill_between(t_holdout, ci_low, ci_high, alpha=0.20, label="IC 95%")
ax.set_title("Real vs Pred — SARIMAX (baseline)")
ax.set_xlabel("Tempo"); ax.set_ylabel("Taxa de desocupação (p.p.)")
ax.legend(); plt.tight_layout(); plt.show(); plt.close(fig)

# -------- Diagnóstico de resíduos --------
resid = sarimax.resid.dropna()

fig2, ax = plt.subplots(1,2, figsize=(9,3.2))
plot_acf(resid.values, lags=16, ax=ax[0]); ax[0].set_title("ACF — Resíduos (SARIMAX)")
plot_pacf(resid.values, lags=16, method="ywm", ax=ax[1]); ax[1].set_title("PACF — Resíduos (SARIMAX)")
plt.tight_layout(); plt.show(); plt.close(fig2)

lb = acorr_ljungbox(resid, lags=[1,4,8,12,16], return_df=True)
display(lb)

jb_stat, jb_p = jarque_bera(resid)
from pandas import Series
skew = float(Series(resid).skew()); kurt = float(Series(resid).kurtosis())
print(f"Jarque-Bera: estat={jb_stat:.3f}, p={jb_p:.4f}, skew={skew:.3f}, kurt(excesso)={kurt:.3f}")

# Teste ARCH-LM (heterocedasticidade condicional) — usa 'nlags' (sem FutureWarning)
arch_stat, arch_p, _, _ = het_arch(resid, nlags=16)
print(f"ARCH-LM (lag 16): p-valor = {arch_p:.4f}  (ideal > 0.05)")

# -------- Comparação com LGBM --------
try:
    comparativo = pd.concat([lgbm_metrics, sarimax_metrics]).round(4)
    display(comparativo)
    fig, ax = plt.subplots(figsize=(7,5))
    ax.plot(t_holdout, y_te.values,          label="Real")
    ax.plot(t_holdout, pred_lgb,             label="LGBM (calibrado)")
    ax.plot(t_holdout, pred_sar.values,      label="SARIMAX")
    ax.set_title("Real vs Pred — Holdout trimestral (últimos 4 trimestres)")
    ax.set_xlabel("Tempo"); ax.set_ylabel("Taxa de desocupação (p.p.)")
    ax.legend(); plt.tight_layout(); plt.show(); plt.close(fig)
except Exception:
    print("Rode LGBM antes deste")
```

```python
import warnings
warnings.filterwarnings(
    "ignore",
    message="Non-invertible starting MA parameters found",
    category=UserWarning,
    module="statsmodels"
)

import numpy as np, pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error

ORDER = (1,1,2)
SEAS_ORDER = (1,0,1,4)   # sazonalidade trimestral (s=4)

def _rmse(a, b):
    a = np.asarray(a, float); b = np.asarray(b, float)
    return np.sqrt(np.mean((a - b)**2))

def rolling_origin_cv(y, X, order=ORDER, seas_order=SEAS_ORDER,
                      initial=36, h=1, step=1, standardize=True):
    """
    Walk-forward (expanding window). Treina até t e prevê t+1..t+h.
    Ajuste com enforce_stationarity/invertibility = True e cov_type='oim'.
    """
    y = y.asfreq("Q-DEC"); X = X.asfreq("Q-DEC")
    n = len(y)
    assert n > initial + h, "Amostra insuficiente para CV (aumente 'initial' ou reduza 'h')."

    rows = []
    fold = 0
    for t in range(initial, n - h + 1, step):
        fold += 1
        y_tr, y_te = y.iloc[:t],      y.iloc[t:t+h]
        X_tr, X_te = X.iloc[:t, :],   X.iloc[t:t+h, :]

        # Padroniza exógenas no TREINO
        if standardize:
            mu = X_tr.mean()
            sd = X_tr.std(ddof=0).replace(0, 1.0)
            X_tr_, X_te_ = (X_tr - mu)/sd, (X_te - mu)/sd
        else:
            X_tr_, X_te_ = X_tr, X_te

        # Ajuste com restrições e fallback de convergência
        model = SARIMAX(y_tr, exog=X_tr_, order=order, seasonal_order=seas_order,
                        enforce_stationarity=True, enforce_invertibility=True)
        try:
            res = model.fit(method="lbfgs", disp=False, maxiter=1000, cov_type="oim")
            if hasattr(res, "mle_retvals") and not res.mle_retvals.get("converged", True):
                raise RuntimeError("fold not converged")
        except Exception:
            # Powell para start, depois L-BFGS novamente
            try:
                res0 = model.fit(method="powell", maxiter=300, disp=False)
                res  = model.fit(method="lbfgs", start_params=res0.params,
                                 disp=False, maxiter=1500, cov_type="oim")
            except Exception:
                # Relaxa constraints para start e reimpõe
                relax = SARIMAX(y_tr, exog=X_tr_, order=order, seasonal_order=seas_order,
                                enforce_stationarity=False, enforce_invertibility=False
                               ).fit(method="lbfgs", disp=False, maxiter=800, cov_type="oim")
                res = SARIMAX(y_tr, exog=X_tr_, order=order, seasonal_order=seas_order,
                              enforce_stationarity=True, enforce_invertibility=True
                             ).fit(method="lbfgs", start_params=relax.params,
                                   disp=False, maxiter=1500, cov_type="oim")

        pred = res.get_forecast(steps=h, exog=X_te_).predicted_mean
        y_hat = np.asarray(pred.values, float).ravel()
        y_ref = np.asarray(y_te.values,  float).ravel()

        rows.append({
            "fold": fold,
            "inicio_treino": y_tr.index[0], "fim_treino": y_tr.index[-1],
            "inicio_teste": y_te.index[0],  "fim_teste": y_te.index[-1],
            "h": h,
            "MAE": mean_absolute_error(y_ref, y_hat),
            "RMSE": _rmse(y_ref, y_hat),
        })

    cv = pd.DataFrame(rows)
    resumo = pd.DataFrame({
        "MAE (média ± dp)":  [f"{cv['MAE'].mean():.3f} ± {cv['MAE'].std(ddof=1):.3f}"],
        "RMSE (média ± dp)": [f"{cv['RMSE'].mean():.3f} ± {cv['RMSE'].std(ddof=1):.3f}"],
        "Folds": [len(cv)],
        "h": [h],
    }, index=[f"SARIMAX (exógenas = {list(X.columns)})"])
    return cv, resumo

y_all    = df_model["tx_desocupacao"]
X_caged  = df_model[["caged_saldo_tri"]]

# Horizonte 1 e resumo
cv_h1, resumo_h1 = rolling_origin_cv(y_all, X_caged, initial=36, h=1, step=1)
display(cv_h1.head(8))
display(resumo_h1)

cv_h4, resumo_h4 = rolling_origin_cv(y_all, X_caged, initial=36, h=4, step=1)
display(resumo_h4)
```

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.stats.diagnostic import acorr_ljungbox
from sklearn.metrics import mean_absolute_error

ORDER = (1,1,2)
SEAS_ORDER = (1,0,1,4)
TEST_LAST_N = 4

assert "selic_tri" in df_model.columns, "Coluna 'selic_tri' não encontrada no df_model (ver dataset tratado)."

train, test = df_model.iloc[:-TEST_LAST_N], df_model.iloc[-TEST_LAST_N:]

y_tr = train["tx_desocupacao"].asfreq("Q-DEC")
y_te = test ["tx_desocupacao"].asfreq("Q-DEC")

X_tr_caged = train[["caged_saldo_tri"]].asfreq("Q-DEC")
X_te_caged = test [["caged_saldo_tri"]].asfreq("Q-DEC")

X_tr_both  = train[["caged_saldo_tri","selic_tri"]].asfreq("Q-DEC")
X_te_both  = test [["caged_saldo_tri","selic_tri"]].asfreq("Q-DEC")

def _rmse(a, b):
    a = np.asarray(a, float); b = np.asarray(b, float)
    return np.sqrt(np.mean((a - b)**2))

def _fit_predict(y_tr, X_tr, y_te, X_te):
    # padroniza no treino
    mu, sd = X_tr.mean(), X_tr.std(ddof=0).replace(0, 1.0)
    Xtr, Xte = (X_tr - mu)/sd, (X_te - mu)/sd

    mod = SARIMAX(y_tr, exog=Xtr, order=ORDER, seasonal_order=SEAS_ORDER,
                  enforce_stationarity=True, enforce_invertibility=True)
    try:
        res = mod.fit(method="lbfgs", disp=False, maxiter=2000, cov_type="oim")
        if hasattr(res, "mle_retvals") and not res.mle_retvals.get("converged", True):
            raise RuntimeError("not converged")
    except Exception:
        # Powell para start -> L-BFGS
        res0 = mod.fit(method="powell", maxiter=300, disp=False)
        res  = mod.fit(method="lbfgs", start_params=res0.params, disp=False, maxiter=3000, cov_type="oim")

    pred = res.get_forecast(steps=len(y_te), exog=Xte).predicted_mean
    return res, pred

# Baseline: CAGED
res_base, pred_base = _fit_predict(y_tr, X_tr_caged, y_te, X_te_caged)

# Ablação: CAGED + SELIC
res_both, pred_both = _fit_predict(y_tr, X_tr_both,  y_te, X_te_both)

# Métricas no holdout
y_true = np.asarray(y_te.values, float).ravel()
m_base = {"MAE": mean_absolute_error(y_true, np.asarray(pred_base.values, float).ravel()),
          "RMSE": _rmse(y_true, np.asarray(pred_base.values, float).ravel())}
m_both = {"MAE": mean_absolute_error(y_true, np.asarray(pred_both.values, float).ravel()),
          "RMSE": _rmse(y_true, np.asarray(pred_both.values, float).ravel())}

comparativo = pd.DataFrame([m_base, m_both],
                           index=["SARIMAX (CAGED)", "SARIMAX (CAGED+SELIC)"]).round(4)
display(comparativo)

# Diagnóstico formal: Ljung-Box em lags curtos e médios
def _lb(resid):
    return acorr_ljungbox(resid.dropna(), lags=[1,4,8,12,16], return_df=True)[["lb_stat","lb_pvalue"]]

print("\nLjung-Box — SARIMAX (CAGED):")
display(_lb(res_base.resid))

print("Ljung-Box — SARIMAX (CAGED+SELIC):")
display(_lb(res_both.resid))

# Gráfico comparando previsões no holdout
t_idx = y_te.index.to_timestamp() if isinstance(y_te.index, pd.PeriodIndex) else y_te.index
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(t_idx, y_te.values,        label="Real")
ax.plot(t_idx, pred_base.values,   label="SARIMAX (CAGED)")
ax.plot(t_idx, pred_both.values,   label="SARIMAX (CAGED+SELIC)")
ax.set_title("Real vs Pred — Holdout (SARIMAX baseline vs +SELIC)")
ax.set_xlabel("Tempo"); ax.set_ylabel("Taxa de desocupação (p.p.)")
ax.legend(); plt.tight_layout(); plt.show(); plt.close(fig)
```

### **Conclusão geral sobre o SARIMAX**

A etapa de modelagem apresentou desafios técnicos principalmente ligados à convergência dos parâmetros e à interpretação dos resultados em amostras curtas.
Na implementação inicial do SARIMAX, surgiram avisos de não-convergência (“Maximum Likelihood optimization failed to converge”).
A solução envolveu a adoção de um processo em duas fases: primeiro, o método Powell para gerar parâmetros iniciais robustos; depois, a otimização L-BFGS com cov_type="oim" e sem o argumento obsoleto tol.
Esse ajuste eliminou os warnings e estabilizou o processo de estimação.

Outro desafio ocorreu na verificação das raízes AR e MA, que inicialmente retornavam valores muito próximos de 1, sinalizando instabilidade numérica.
Foi necessário revisar as ordens sazonais (reduzindo Q de 2 para 1) e reforçar as restrições enforce_stationarity=True e enforce_invertibility=True.
Após essa correção, o modelo tornou-se estacionário e invertível.

Nos diagnósticos de resíduos, o Jarque–Bera indicou leve leptocurtose (caudas pesadas), mas os testes Ljung-Box e ARCH-LM confirmaram ausência de autocorrelação e heterocedasticidade.
Optou-se por manter cov_type="oim", com a observação de que erros-padrão robustos (robust_oim) podem ser empregados em inferências futuras.

Já o modelo de Machine Learning (LGBM) enfrentou dificuldade para capturar viradas de curto prazo devido ao tamanho limitado da amostra e ao pequeno conjunto de features.
Mesmo após calibração de viés, as previsões permaneceram planas e superestimadas, o que evidenciou a limitação dos modelos de boosting em séries pequenas e fortemente sazonais.

Em contrapartida, o SARIMAX apresentou alta aderência visual e erro médio inferior a 0,3 p.p., superando todos os benchmarks.
O aprendizado obtido com esses ajustes consolidou um pipeline estável e reprodutível, pronto para refinamentos na Etapa 4

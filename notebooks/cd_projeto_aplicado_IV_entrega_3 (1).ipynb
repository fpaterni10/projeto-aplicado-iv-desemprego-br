# **Referencial Te√≥rico**

As s√©ries temporais constituem uma classe de dados onde a depend√™ncia temporal √© o principal fator de an√°lise, permitindo identificar tend√™ncias, ciclos e padr√µes sazonais (ENDERS, 2015). Os modelos tradicionais para lidar com esse tipo de dado incluem o ARIMA (AutoRegressive Integrated Moving Average), que combina componentes autoregressivos (AR), de integra√ß√£o (I) e de m√©dias m√≥veis (MA). Entretanto, quando as s√©ries apresentam padr√µes sazonais, √© necess√°rio recorrer a modelos estendidos como o SARIMA, que incorpora a sazonalidade, e o SARIMAX, que permite a inclus√£o de vari√°veis ex√≥genas relevantes, como indicadores macroecon√¥micos (BEIR√ÉO; GON√áALVES; NETO, 2021).

Um ponto central em s√©ries temporais √© a estacionariedade, que ocorre quando a m√©dia e a vari√¢ncia permanecem constantes ao longo do tempo. S√©ries com forte tend√™ncia ou sazonalidade n√£o s√£o estacion√°rias e precisam ser transformadas, normalmente por diferencia√ß√£o ou aplica√ß√£o de logaritmos. Isso √© fundamental para que m√©todos como ARIMA e SARIMA sejam aplic√°veis de forma adequada (ENDERS, 2015).

No contexto do desemprego, diversos trabalhos no Brasil t√™m demonstrado a relev√¢ncia da modelagem estat√≠stica. Becker (2010), em estudo sobre a taxa de desemprego da Regi√£o Metropolitana de Porto Alegre, avaliou a aplica√ß√£o de modelos como Holt-Winters e SARIMA, concluindo que modelos sazonais oferecem melhor ajuste em s√©ries com ciclos econ√¥micos regulares. J√° Beir√£o, Gon√ßalves e Neto (2021), ao analisar a taxa de desemprego brasileira com a metodologia Box-Jenkins, confirmaram a necessidade de diferencia√ß√£o para lidar com a n√£o estacionariedade da s√©rie.

Al√©m dos m√©todos estat√≠sticos cl√°ssicos, h√° uma crescente utiliza√ß√£o de algoritmos de aprendizado de m√°quina. Trabalhos recentes apontam que abordagens h√≠bridas, combinando modelos de s√©ries temporais com machine learning, apresentam maior robustez e menor erro de previs√£o. Um exemplo √© o estudo publicado em Estudos Econ√¥micos (2020), que utilizou modelos fatoriais de alta dimens√£o aliados a t√©cnicas de sele√ß√£o de vari√°veis para prever indicadores macroecon√¥micos, incluindo a taxa de desemprego, alcan√ßando ganhos significativos de precis√£o.

A literatura mostra, portanto, que a modelagem de s√©ries temporais voltada para a previs√£o de indicadores de mercado de trabalho no Brasil √© um tema relevante, com aplica√ß√µes pr√°ticas em pol√≠ticas p√∫blicas e gest√£o econ√¥mica. O uso combinado de t√©cnicas estat√≠sticas cl√°ssicas e m√©todos modernos de aprendizado de m√°quina surge como alternativa promissora, unindo a interpretabilidade dos primeiros √† flexibilidade dos segundos.

# **EDA e Pr√©-processamento dos dados**

Foram exploradas e analisadas as bases de dados PNAD Cont√≠nua (tabela 4099) e CAGED, abrangendo o per√≠odo de 2012 a 2025.
A consolida√ß√£o das fontes teve como objetivo garantir consist√™ncia temporal trimestral, condi√ß√£o essencial para aplica√ß√£o de modelos de s√©ries temporais como SARIMAX e LGBM Regressor.

Durante o pr√©-processamento, foram removidos valores ausentes e outliers, aplicadas t√©cnicas de preenchimento (`ffill`) e criadas vari√°veis derivadas, como:

- `caged_roll3`: m√©dia m√≥vel trimestral do saldo de empregos;
- `caged_roll3_asinh`: transforma√ß√£o hiperb√≥lica para estabilizar vari√¢ncia;
- dummies de trimestre e do regime p√≥s-2021, representando mudan√ßas estruturais nas s√©ries.

**Conclus√£o da EDA.** As an√°lises confirmaram sazonalidade trimestral e estacionariedade parcial, justificando o uso combinado de **SARIMAX (1,1,2)√ó(1,0,1,4)** e **LGBM Regressor**.


```

#@title üîß Setup & Carregamento (execute antes dos modelos)
import warnings, numpy as np, pandas as pd, random
warnings.filterwarnings("ignore")
np.random.seed(42); random.seed(42)

RAW_URL = (
    "https://raw.githubusercontent.com/"
    "fpaterni10/projeto-aplicado-iv-desemprego-br/main/"
    "dataset/tratados/caged_pnad_selic_trimestral.csv"
)

# Carrega base trimestral consolidada
df0 = pd.read_csv(RAW_URL)
df0["periodo_tri"] = df0["periodo_tri"].astype(str)

# √çndice trimestral com frequ√™ncia expl√≠cita (Q-DEC) e coluna 'data'
per = pd.PeriodIndex(df0["periodo_tri"], freq="Q-DEC")
df = df0.set_index(per).sort_index()
df.index.name = "periodo"
df["data"] = df.index.to_timestamp(how="end").normalize()

# Features b√°sicas
df["caged_roll3"] = df["caged_saldo_tri"].rolling(3, min_periods=3).mean()
df["caged_roll3_asinh"] = np.arcsinh(df["caged_roll3"])

# Dummies sazonais + flag de regime p√≥s-2021
df["quarter"] = df.index.quarter
Q = pd.get_dummies(df["quarter"], prefix="Q", drop_first=False)
df["post_2021"] = (df["data"] >= "2021-01-01").astype(int)

# Base de modelagem (remove NaNs iniciais de rolling)
df_model = pd.concat([df, Q], axis=1).dropna().copy()

# ML: lags, deltas e volatilidade curtinha
for k in [1, 2, 3, 4]:
    df_model[f"tx_lag{k}"] = df_model["tx_desocupacao"].shift(k)
df_model["caged_diff1"] = df_model["caged_saldo_tri"].diff(1)
df_model["selic_diff1"] = df_model["selic_tri"].diff(1)
df_model["caged_roll3_std"] = df_model["caged_saldo_tri"].rolling(3).std()

# Limpeza final p√≥s-lags/diffs/std
df_model = df_model.dropna().copy()

# Diagn√≥stico r√°pido
print("Dimens√£o df_model:", df_model.shape)
df_model.info()
```

```

#@title üìé Utilit√°rios: m√©tricas e split temporal
import numpy as np, pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error

def temporal_holdout(df, n_test=4):
    """√öltimos n_test trimestres = teste, preservando a ordem temporal."""
    if "data" in df.columns:
        df2 = df.sort_values("data").copy()
    else:
        df2 = df.sort_index().copy()
    return df2.iloc[:-n_test].copy(), df2.iloc[-n_test:].copy()

def metrics_report(y_true, y_pred, prefix="modelo"):
    """MAE, RMSE e MAPE (%) ‚Äî sem usar kwargs incompat√≠veis (ex.: squared=False em vers√µes antigas)."""
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    n = min(len(y_true), len(y_pred))
    y_true = y_true[:n]; y_pred = y_pred[:n]
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    denom = np.where(np.abs(y_true) < 1e-12, np.nan, y_true)
    mape = np.nanmean(np.abs((y_true - y_pred) / denom)) * 100
    return pd.DataFrame({"MAE":[mae], "RMSE":[rmse], "MAPE (%)":[mape]}, index=[prefix])
```


## **Modelo 1 ‚Äî LGBM (Machine Learning)**

O **LightGBM Regressor** √© treinado com lags do alvo, vari√°veis derivadas do CAGED/SELIC e dummies sazonais.
A valida√ß√£o utiliza *holdout* nos √∫ltimos 4 trimestres.


```

#@title ü§ñ LGBM Regressor ‚Äî treino e avalia√ß√£o
from lightgbm import LGBMRegressor
import contextlib, io, os

if "df_model" not in globals():
    raise RuntimeError("Execute a c√©lula 'Setup & Carregamento' antes dos modelos.")

# Conjunto de features (mant√©m apenas as existentes)
features = [
    "caged_roll3_asinh", "caged_saldo_tri", "caged_roll3",
    "caged_diff1", "caged_roll3_std",
    "selic_tri", "selic_diff1",
    "tx_lag1","tx_lag2","tx_lag3","tx_lag4",
    "Q_1","Q_2","Q_3","Q_4","post_2021"
]
features = [f for f in features if f in df_model.columns]
target = "tx_desocupacao"

# Split temporal (√∫ltimos 4 trimestres = teste)
train_lgb, test_lgb = temporal_holdout(df_model, n_test=4)
X_train, y_train = train_lgb[features], train_lgb[target].to_numpy()
X_test,  y_test  = test_lgb[features],  test_lgb[target].to_numpy()

# Modelo (silencioso, sem logs ou avisos de par√¢metros conflitantes)
lgbm = LGBMRegressor(
    n_estimators=800, learning_rate=0.03,
    num_leaves=63, max_depth=6,
    min_child_samples=5,              # use APENAS este (evita aviso com min_data_in_leaf)
    subsample=0.9, colsample_bytree=0.9,
    reg_lambda=0.1,
    random_state=42, n_jobs=-1, verbosity=-1
)

with contextlib.redirect_stdout(io.StringIO()), contextlib.redirect_stderr(io.StringIO()):
    lgbm.fit(X_train, y_train)

pred_lgb = lgbm.predict(X_test)
lgbm_metrics = metrics_report(y_test, pred_lgb, prefix="LGBM (seed=42)")
lgbm_metrics
```


## **Modelo 2 ‚Äî SARIMAX (Estat√≠stico)**

O **SARIMAX (1, 1, 2) √ó (1, 0, 1, 4)** √© ajustado com vari√°veis ex√≥genas `caged_saldo_tri` e `caged_roll3_asinh`.
A frequ√™ncia trimestral expl√≠cita evita *ValueWarning* de √≠ndice e *shape mismatch* no *forecast*.


```

#@title üìà SARIMAX ‚Äî treino e avalia√ß√£o
from statsmodels.tsa.statespace.sarimax import SARIMAX

if "df_model" not in globals():
    raise RuntimeError("Execute a c√©lula 'Setup & Carregamento' antes dos modelos.")

# Ex√≥genas
exog_cols = ["caged_saldo_tri", "caged_roll3_asinh"]
exog_cols = [c for c in exog_cols if c in df_model.columns]

# Split temporal id√™ntico ao do LGBM
train_sar, test_sar = temporal_holdout(df_model, n_test=4)

# End√≥gena e ex√≥genas com MESMA frequ√™ncia trimestral
y_train = train_sar["tx_desocupacao"].asfreq("Q-DEC")
y_test  = test_sar["tx_desocupacao"].asfreq("Q-DEC")
X_train = train_sar[exog_cols].asfreq("Q-DEC")
X_test  = test_sar[exog_cols].asfreq("Q-DEC")

# Ajuste do SARIMAX
sarimax = SARIMAX(
    endog=y_train, exog=X_train,
    order=(1,1,2), seasonal_order=(1,0,1,4),
    enforce_stationarity=False, enforce_invertibility=False
).fit(disp=False, maxiter=500)

# Forecast fora da amostra com ex√≥genas alinhadas
pred_sar = sarimax.get_forecast(steps=len(y_test), exog=X_test).predicted_mean

sarimax_metrics = metrics_report(y_test.values, pred_sar.values, prefix="SARIMAX (1,1,2)(1,0,1,4)")
sarimax_metrics
```


## **Compara√ß√£o entre Modelos e Gr√°fico (Holdout de 4 trimestres)**


```

#@title üîé Comparativo de m√©tricas + gr√°fico Real vs Pred
import matplotlib.pyplot as plt
import os

# Tabela comparativa
comparativo = pd.concat([lgbm_metrics, sarimax_metrics])
display(comparativo)

# Eixo temporal do holdout
t = test_lgb["data"] if "data" in test_lgb.columns else test_lgb.index.to_timestamp()

# Gr√°fico
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(t, y_test,   label="Real")
ax.plot(t, pred_lgb, label="LGBM")
ax.plot(t, pred_sar, label="SARIMAX")
ax.set_title("Real vs Pred ‚Äî Holdout trimestral (√∫ltimos 4 trimestres)")
ax.legend()
fig.tight_layout()

# Salvar figura (para uso em README/relat√≥rio)
os.makedirs("docs/figuras", exist_ok=True)
fig.savefig("docs/figuras/real_vs_pred_holdout.png", dpi=160, bbox_inches="tight")

plt.show()
plt.close(fig)
```

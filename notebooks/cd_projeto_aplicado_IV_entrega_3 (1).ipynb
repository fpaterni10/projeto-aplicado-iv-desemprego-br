# **Referencial Teórico**

As séries temporais constituem uma classe de dados onde a dependência temporal é o principal fator de análise, permitindo identificar tendências, ciclos e padrões sazonais (ENDERS, 2015). Os modelos tradicionais para lidar com esse tipo de dado incluem o ARIMA (AutoRegressive Integrated Moving Average), que combina componentes autoregressivos (AR), de integração (I) e de médias móveis (MA). Entretanto, quando as séries apresentam padrões sazonais, é necessário recorrer a modelos estendidos como o SARIMA, que incorpora a sazonalidade, e o SARIMAX, que permite a inclusão de variáveis exógenas relevantes, como indicadores macroeconômicos (BEIRÃO; GONÇALVES; NETO, 2021).

Um ponto central em séries temporais é a estacionariedade, que ocorre quando a média e a variância permanecem constantes ao longo do tempo. Séries com forte tendência ou sazonalidade não são estacionárias e precisam ser transformadas, normalmente por diferenciação ou aplicação de logaritmos. Isso é fundamental para que métodos como ARIMA e SARIMA sejam aplicáveis de forma adequada (ENDERS, 2015).

No contexto do desemprego, diversos trabalhos no Brasil têm demonstrado a relevância da modelagem estatística. Becker (2010), em estudo sobre a taxa de desemprego da Região Metropolitana de Porto Alegre, avaliou a aplicação de modelos como Holt-Winters e SARIMA, concluindo que modelos sazonais oferecem melhor ajuste em séries com ciclos econômicos regulares. Já Beirão, Gonçalves e Neto (2021), ao analisar a taxa de desemprego brasileira com a metodologia Box-Jenkins, confirmaram a necessidade de diferenciação para lidar com a não estacionariedade da série.

Além dos métodos estatísticos clássicos, há uma crescente utilização de algoritmos de aprendizado de máquina. Trabalhos recentes apontam que abordagens híbridas, combinando modelos de séries temporais com machine learning, apresentam maior robustez e menor erro de previsão. Um exemplo é o estudo publicado em Estudos Econômicos (2020), que utilizou modelos fatoriais de alta dimensão aliados a técnicas de seleção de variáveis para prever indicadores macroeconômicos, incluindo a taxa de desemprego, alcançando ganhos significativos de precisão.

A literatura mostra, portanto, que a modelagem de séries temporais voltada para a previsão de indicadores de mercado de trabalho no Brasil é um tema relevante, com aplicações práticas em políticas públicas e gestão econômica. O uso combinado de técnicas estatísticas clássicas e métodos modernos de aprendizado de máquina surge como alternativa promissora, unindo a interpretabilidade dos primeiros à flexibilidade dos segundos.

# **EDA e Pré-processamento dos dados**

Foram exploradas e analisadas as bases de dados PNAD Contínua (tabela 4099) e CAGED, abrangendo o período de 2012 a 2025.
A consolidação das fontes teve como objetivo garantir consistência temporal trimestral, condição essencial para aplicação de modelos de séries temporais como SARIMAX e LGBM Regressor.

Durante o pré-processamento, foram removidos valores ausentes e outliers, aplicadas técnicas de preenchimento (`ffill`) e criadas variáveis derivadas, como:

- `caged_roll3`: média móvel trimestral do saldo de empregos;
- `caged_roll3_asinh`: transformação hiperbólica para estabilizar variância;
- dummies de trimestre e do regime pós-2021, representando mudanças estruturais nas séries.

**Conclusão da EDA.** As análises confirmaram sazonalidade trimestral e estacionariedade parcial, justificando o uso combinado de **SARIMAX (1,1,2)×(1,0,1,4)** e **LGBM Regressor**.


```

#@title 🔧 Setup & Carregamento (execute antes dos modelos)
import warnings, numpy as np, pandas as pd, random
warnings.filterwarnings("ignore")
np.random.seed(42); random.seed(42)

RAW_URL = (
    "https://raw.githubusercontent.com/"
    "fpaterni10/projeto-aplicado-iv-desemprego-br/main/"
    "dataset/tratados/caged_pnad_selic_trimestral.csv"
)

# Carrega base trimestral consolidada
df0 = pd.read_csv(RAW_URL)
df0["periodo_tri"] = df0["periodo_tri"].astype(str)

# Índice trimestral com frequência explícita (Q-DEC) e coluna 'data'
per = pd.PeriodIndex(df0["periodo_tri"], freq="Q-DEC")
df = df0.set_index(per).sort_index()
df.index.name = "periodo"
df["data"] = df.index.to_timestamp(how="end").normalize()

# Features básicas
df["caged_roll3"] = df["caged_saldo_tri"].rolling(3, min_periods=3).mean()
df["caged_roll3_asinh"] = np.arcsinh(df["caged_roll3"])

# Dummies sazonais + flag de regime pós-2021
df["quarter"] = df.index.quarter
Q = pd.get_dummies(df["quarter"], prefix="Q", drop_first=False)
df["post_2021"] = (df["data"] >= "2021-01-01").astype(int)

# Base de modelagem (remove NaNs iniciais de rolling)
df_model = pd.concat([df, Q], axis=1).dropna().copy()

# ML: lags, deltas e volatilidade curtinha
for k in [1, 2, 3, 4]:
    df_model[f"tx_lag{k}"] = df_model["tx_desocupacao"].shift(k)
df_model["caged_diff1"] = df_model["caged_saldo_tri"].diff(1)
df_model["selic_diff1"] = df_model["selic_tri"].diff(1)
df_model["caged_roll3_std"] = df_model["caged_saldo_tri"].rolling(3).std()

# Limpeza final pós-lags/diffs/std
df_model = df_model.dropna().copy()

# Diagnóstico rápido
print("Dimensão df_model:", df_model.shape)
df_model.info()
```

```

#@title 📎 Utilitários: métricas e split temporal
import numpy as np, pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error

def temporal_holdout(df, n_test=4):
    """Últimos n_test trimestres = teste, preservando a ordem temporal."""
    if "data" in df.columns:
        df2 = df.sort_values("data").copy()
    else:
        df2 = df.sort_index().copy()
    return df2.iloc[:-n_test].copy(), df2.iloc[-n_test:].copy()

def metrics_report(y_true, y_pred, prefix="modelo"):
    """MAE, RMSE e MAPE (%) — sem usar kwargs incompatíveis (ex.: squared=False em versões antigas)."""
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    n = min(len(y_true), len(y_pred))
    y_true = y_true[:n]; y_pred = y_pred[:n]
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    denom = np.where(np.abs(y_true) < 1e-12, np.nan, y_true)
    mape = np.nanmean(np.abs((y_true - y_pred) / denom)) * 100
    return pd.DataFrame({"MAE":[mae], "RMSE":[rmse], "MAPE (%)":[mape]}, index=[prefix])
```


## **Modelo 1 — LGBM (Machine Learning)**

O **LightGBM Regressor** é treinado com lags do alvo, variáveis derivadas do CAGED/SELIC e dummies sazonais.
A validação utiliza *holdout* nos últimos 4 trimestres.


```

#@title 🤖 LGBM Regressor — treino e avaliação
from lightgbm import LGBMRegressor
import contextlib, io, os

if "df_model" not in globals():
    raise RuntimeError("Execute a célula 'Setup & Carregamento' antes dos modelos.")

# Conjunto de features (mantém apenas as existentes)
features = [
    "caged_roll3_asinh", "caged_saldo_tri", "caged_roll3",
    "caged_diff1", "caged_roll3_std",
    "selic_tri", "selic_diff1",
    "tx_lag1","tx_lag2","tx_lag3","tx_lag4",
    "Q_1","Q_2","Q_3","Q_4","post_2021"
]
features = [f for f in features if f in df_model.columns]
target = "tx_desocupacao"

# Split temporal (últimos 4 trimestres = teste)
train_lgb, test_lgb = temporal_holdout(df_model, n_test=4)
X_train, y_train = train_lgb[features], train_lgb[target].to_numpy()
X_test,  y_test  = test_lgb[features],  test_lgb[target].to_numpy()

# Modelo (silencioso, sem logs ou avisos de parâmetros conflitantes)
lgbm = LGBMRegressor(
    n_estimators=800, learning_rate=0.03,
    num_leaves=63, max_depth=6,
    min_child_samples=5,              # use APENAS este (evita aviso com min_data_in_leaf)
    subsample=0.9, colsample_bytree=0.9,
    reg_lambda=0.1,
    random_state=42, n_jobs=-1, verbosity=-1
)

with contextlib.redirect_stdout(io.StringIO()), contextlib.redirect_stderr(io.StringIO()):
    lgbm.fit(X_train, y_train)

pred_lgb = lgbm.predict(X_test)
lgbm_metrics = metrics_report(y_test, pred_lgb, prefix="LGBM (seed=42)")
lgbm_metrics
```


## **Modelo 2 — SARIMAX (Estatístico)**

O **SARIMAX (1, 1, 2) × (1, 0, 1, 4)** é ajustado com variáveis exógenas `caged_saldo_tri` e `caged_roll3_asinh`.
A frequência trimestral explícita evita *ValueWarning* de índice e *shape mismatch* no *forecast*.


```

#@title 📈 SARIMAX — treino e avaliação
from statsmodels.tsa.statespace.sarimax import SARIMAX

if "df_model" not in globals():
    raise RuntimeError("Execute a célula 'Setup & Carregamento' antes dos modelos.")

# Exógenas
exog_cols = ["caged_saldo_tri", "caged_roll3_asinh"]
exog_cols = [c for c in exog_cols if c in df_model.columns]

# Split temporal idêntico ao do LGBM
train_sar, test_sar = temporal_holdout(df_model, n_test=4)

# Endógena e exógenas com MESMA frequência trimestral
y_train = train_sar["tx_desocupacao"].asfreq("Q-DEC")
y_test  = test_sar["tx_desocupacao"].asfreq("Q-DEC")
X_train = train_sar[exog_cols].asfreq("Q-DEC")
X_test  = test_sar[exog_cols].asfreq("Q-DEC")

# Ajuste do SARIMAX
sarimax = SARIMAX(
    endog=y_train, exog=X_train,
    order=(1,1,2), seasonal_order=(1,0,1,4),
    enforce_stationarity=False, enforce_invertibility=False
).fit(disp=False, maxiter=500)

# Forecast fora da amostra com exógenas alinhadas
pred_sar = sarimax.get_forecast(steps=len(y_test), exog=X_test).predicted_mean

sarimax_metrics = metrics_report(y_test.values, pred_sar.values, prefix="SARIMAX (1,1,2)(1,0,1,4)")
sarimax_metrics
```


## **Comparação entre Modelos e Gráfico (Holdout de 4 trimestres)**


```

#@title 🔎 Comparativo de métricas + gráfico Real vs Pred
import matplotlib.pyplot as plt
import os

# Tabela comparativa
comparativo = pd.concat([lgbm_metrics, sarimax_metrics])
display(comparativo)

# Eixo temporal do holdout
t = test_lgb["data"] if "data" in test_lgb.columns else test_lgb.index.to_timestamp()

# Gráfico
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(t, y_test,   label="Real")
ax.plot(t, pred_lgb, label="LGBM")
ax.plot(t, pred_sar, label="SARIMAX")
ax.set_title("Real vs Pred — Holdout trimestral (últimos 4 trimestres)")
ax.legend()
fig.tight_layout()

# Salvar figura (para uso em README/relatório)
os.makedirs("docs/figuras", exist_ok=True)
fig.savefig("docs/figuras/real_vs_pred_holdout.png", dpi=160, bbox_inches="tight")

plt.show()
plt.close(fig)
```

# **Introdução**

Segundo o IBGE, a taxa de desocupação no Brasil foi de 7,8% em 2023. No estado de São Paulo, a média anual ficou em 7,2%, sinalizando que o mercado de trabalho ainda reflete os efeitos da pandemia.

A oscilação da desocupação afeta diretamente milhões de pessoas e influencia políticas públicas, decisões econômicas e projeções. No contexto de São Paulo, que é a maior economia do país,  prever essa taxa com precisão torna-se estratégico para ações como planejamento orçamentário, atração de investimentos e combate à informalidade.

Este projeto parte do princípio de que, com o uso de modelos estatísticos e séries temporais, é possível antecipar cenários e flutuações da desocupação, permitindo maior previsibilidade para todos os setores impactados por esse indicador.


# **Motivações e justificativa**

A taxa de desocupação é um indicador-chave para o entendimento do mercado de trabalho e da economia. No entanto, mesmo com os dados disponibilizados pelo IBGE, ainda são raros os estudos que utilizam modelos estatísticos para prever a desocupação em nível estadual, como em São Paulo.

Este projeto busca preencher essa lacuna por meio de um modelo simples de previsão com base em séries temporais, permitindo observar tendências e apoiar futuras análises.

# **Descrição da base de dados**

A base de dados utilizada neste projeto foi extraída da Pesquisa Nacional por Amostra de Domicílios Contínua – PNAD Contínua Trimestral, disponibilizada pelo Instituto Brasileiro de Geografia e Estatística (IBGE).

- Formato dos dados: Planilha eletrônica no formato .xlsx (Microsoft Excel).
- Período de coleta: De 2012 (1º trimestre) até 2025 (2º trimestre).
- Intervalo de coleta: Trimestral, com dados representando médias dos três meses de cada trimestre.
- Segmentação: Os dados foram segmentados por Unidade da Federação (UF). Para este projeto, utilizamos exclusivamente os dados do estado de São Paulo.
- Indicador escolhido: Taxa de desocupação da população economicamente ativa.
